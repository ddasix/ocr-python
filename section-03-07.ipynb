```python
# Thresholding
import cv2
from pytesseract import Output, pytesseract
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./OCR_with_Python/Images/page-book.jpg')
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
value, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
value
```

```python
value, thresh = cv2.threshold(gray_img, 180, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
# Otsu Method
value, otsu = cv2.threshold(gray_img, 140, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt.imshow(otsu, cmap='gray')
```

```python
# Resizing
#       INTER_NEAREST - a nearest neighbor interpolation. It is widely used because it is the fastest
#       INTER_LINEAR - a bilinear interpolation (it's used by default), generally good for zooming in and out of images
#       INTER_AREA - uses the pixel area ratio. May be a preferred method for image reduction as it provides good results
#       INTER_CUBIC - bicubic (4x4 neighboring pixels). It has better results
#       INTER_LANCZOS4 - Lanczos interpolation (8x8 neighboring pixels). Among these algorithms, it is the one with the best quality results.
```

```python
plt.imshow(gray_img, cmap='gray')
print(gray_img.shape)
```

```python
increase_img = cv2.resize(gray_img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
plt.imshow(increase_img, cmap='gray')
print(increase_img.shape)
```

```python
decrease_img = cv2.resize(gray_img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
plt.imshow(decrease_img, cmap='gray')
print(decrease_img.shape)
```

```python
# Thresholding
import cv2
from pytesseract import Output, pytesseract
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./OCR_with_Python/Images/page-book.jpg')
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
value, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
value
```

```python
value, thresh = cv2.threshold(gray_img, 180, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
value, thresh = cv2.threshold(gray_img, 140, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
# Otsu Method
value, otsu = cv2.threshold(gray_img, 140, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt.imshow(otsu, cmap='gray')
```

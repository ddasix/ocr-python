```python
# Thresholding
import cv2
from pytesseract import Output, pytesseract
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    # fig = plt.figure(figsize=(11.69,8.27)) # A4 사이즈
    fig = plt.figure(figsize=(16.53, 11.69))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('./OCR_with_Python/Images/test_noise.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(gray_img)
```

```python
average_blur_img = cv2.blur(gray_img, (5,5))
display(average_blur_img)
```

```python
# Gaussian Blur
gaussian_blur_img = cv2.GaussianBlur(gray_img, (5,5), sigmaX=0)
display(gaussian_blur_img)
```

```python
# Median Blur
median_blur_img = cv2.medianBlur(gray_img, 3)
display(median_blur_img)
```

```python
# Bilateral Filter
bilateral_filter_img = cv2.bilateralFilter(gray_img, 15, 55, 45)
display(bilateral_filter_img)
```

```python
# Text Detection

config_tesseract = '--tessdata-dir tessdata'
text = pytesseract.image_to_string(average_blur_img, lang='por', config=config_tesseract)
print(text)
```

```python

text = pytesseract.image_to_string(median_blur_img, lang='por', config=config_tesseract)
print(text)
```

```python

text = pytesseract.image_to_string(bilateral_filter_img, lang='por', config=config_tesseract)
print(text)
```
